```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Assignment 1: Working with numpy

```python
#Q1 (a)
arr=np.arange(1,21,1)
arr_sum=np.sum(arr)
print(arr_sum)
arr_mean=np.mean(arr)
print(arr_mean)
arr_median=np.median(arr)
print(arr_median)
arr_std=np.std(arr)
print(arr_std)
```

```python
#Q1 (b)
for i in range (0,20,1):
    if arr[i]>10:
        print(i)
```

```python
#Q2
arr1=np.arange(1,17).reshape(4,4)
arr_T=np.empty((4,4))
#(a)
print(arr1)
#(b)
for i in range (0,4):
    for j in range(0,4):
       arr_T[i,j]=arr1[j,i]
print(arr_T)
#(c)
row_sum=np.sum(arr1, axis=1)
coloumn_sum=np.sum(arr1,axis=0)
print(row_sum)
print(coloumn_sum)
```

```python
#Q3
arr2=np.random.randint(1,20,(3,3))
arr3=np.random.randint(1,20,(3,3))
print(arr2)
print(arr3)
#(a)
for i in range (0,3):
    for j in range (0,3):
        print(arr2[i,j]+arr3[i,j])#Element wise addition
        print(arr2[i,j]*arr3[i,j])#multiplication
        print(arr2[i,j]-arr3[i,j])#substraction
#(b)
print(np.dot(arr2,arr3))
```

```python
#Q4
arr4=np.arange(1,13).reshape(3,4)
print(arr4)
print("First row: ", arr4[0,:])
print("Second row: ", arr4[1,:])
print("last coloumn: ", arr4[:,-1])
print("second last coloumn: ", arr4[:,-2])
```

### Assignment 2: Working with Pandas

```python
#Q1
df=pd.DataFrame({
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
})
#(a)
print(df.head(5))
#(b)
print(df['Age'])
print(df['Salary'])
#(c)
print("Average Salary or HR employees: ",df['Salary'].mean())
```

```python
#Q2
df['Bonus']=df['Salary']*0.1
df
```

```python
#Q3
filtered_df=df[(df['Age']>25) & (df['Age']<30)]
print(filtered_df)
```

```python
#Q4
df.groupby('Department')['Salary'].mean()
```

```python
#Q5
sorted_df=df.sort_values(by='Salary', ascending=True)
sorted_df.to_csv("Sorted DataFrame.csv", index=True)
```

### Assignment 3: Working with Matplotlib

```python
#Q1
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

```python
#Q2
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color=['blue', 'green', 'orange', 'red'])
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#Q3
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1, 0, 0, 0]  # Highlight North America

plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', startangle=90)
plt.title("Revenue Distribution by Region")
plt.show()
```

```python
#Q4
data = np.random.randint(1, 101, 1000)

plt.hist(data, bins=10)
plt.title("Histogram of Random Integers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
